The ast module helps Python applications to process trees of the Python abstract syntax grammar.

literal_eval: Safely evaluates an expression node or a string containing a Python 
              literal(strings,bytes,numbers,tuples,lists,dicts,sets)or container display

A lambda function(anonymous function) can take any number of arguments, but can only have one expression.

The map() function in Python takes in a function and a list as argument. The function is called with a lambda function and a list and a new list is returned which contains all the lambda modified items returned by that function for each item. 

```python
# literal_eval(x) : it actually convert the string into list
#lambda function is applying in that particular literal_eval function
# The power of lambda is better shown when you use them as an anonymous function inside another function.
# In genres their is List of key-value dictionary pair
import ast

movies_merge['cast'] = movies_merge['cast'].map(lambda x: ast.literal_eval(x))
movies_merge['crew'] = movies_merge['crew'].map(lambda x: ast.literal_eval(x))
```

```python
movies_merge['genres']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
movies_merge['actor_1_name'] = movies_merge['cast'].map(lambda x: get_actor1(x))
```

```python
movies_merge['actor_1_name'] 
```

```python
# Here we use len(cast)<=1 because if len is 1 or less than 1 then we only get actor1 name 
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
movies_merge['actor_2_name'] = movies_merge['cast'].map(lambda x: get_actor2(x))
```

```python
movies_merge['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2 :
        return np.NaN
    else:
        return (casts[2])
```

```python
movies_merge['actor_3_name'] =movies_merge['cast'].map(lambda x: get_actor3(x))
```

```python
movies_merge['actor_3_name']
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
movies_merge['director_name'] = movies_merge['crew'].map(lambda x: get_directors(x))
```

```python
movies_merge['director_name']
```

# Using Weighted average for each movie's Average Rating

The following formula is used to calculate the Top Rated 250 titles. This formula provides a true 'Bayesian estimate', which takes into account the number of votes each title has received, minimum votes required to be on the list, and the mean vote for all titles:
            W=(Rv + Cm)/(v + m)
where: W = Weighted Rating

R = average for the movie as a number from 0 to 10(mean)=(Rating)

v = number of votes for the movies = (votes)

m = minimum votes requird to be listed in the Top 250(currently 3000)

C = the mean vote across the whole report(currently 6.9)


# Source: 
https://help.imdb.com/article/imdb/track-movies-tv/ratings-faq/G67Y87TFYYP6TWAV#calculatetop

```python
# Calculate all the components based on the above formula:
v = movie['vote_count']
R = movie['vote_average']
C = movie['vote_average'].mean()
m = movie['vote_count'].quantile(0.70)
```

```python
movie['weighted_average']= ((R*v)+(C*m))/(v+m)
```

```python
movie.head()
```

```python
# Sorting
movie_sorted_ranking = movie.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['movie_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
weight_average = movie_sorted_ranking.sort_values('weighted_average',ascending = False)
plt.figure(figsize =(12,6))
axis1 = sns.barplot(x = weight_average['weighted_average'].head(10), y =weight_average['movie_title'].head())
plt.title('Best Movies by average votes',weight ='bold')
plt.xlabel('Weighted Average score',weight ='bold')
plt.ylabel('Movie Title',weight = 'bold')
plt.savefig('best_movies.png')

popularity = movie_sorted_ranking.sort_values('popularity',ascending = False)
plt.figure(figsize =(12,6))
axis1 = sns.barplot(x = popularity['popularity'].head(10), y =popularity['movie_title'].head())
plt.title('Most Popular by  votes',weight ='bold')
plt.xlabel('Score of Popularity',weight ='bold')
plt.ylabel('Movie Title',weight = 'bold')
plt.savefig('best_popular_movies.png')
```

# Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movie[['weighted_average','popularity']])
movie_normalization_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalization_df.head()
```

```python
movie[['normalized_weight_average','normalized_popularity']]= movie_normalization_df
```

```python
movie.head()
```

```python
movie['score'] = movie['normalized_weight_average']*0.5 +movie['normalized_popularity']*0.5
movies_scored_df = movie.sort_values(['score'],ascending = False)

movies_scored_df[['movie_title','normalized_weight_average','normalized_popularity','score']].head(20)
```

```python
scored_df = movie.sort_values('score',ascending = False)

plt.figure(figsize =(12,6))

axis = sns.barplot(x = scored_df['score'].head(10), y = scored_df['movie_title'].head(10),data = scored_df,palette ='deep')
#plt.xlim(4,10)
plt.title('Best Rated & Most Popular Blend',weight ='bold')
plt.xlabel('Score',weight ='bold')
plt.ylabel('Movie Title',weight = 'bold')
plt.savefig('scoredmovies.png')
```
